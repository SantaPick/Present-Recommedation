## 2. 심리테스트 진행 시뮬레이션


```python
def simulate_psychology_test():
    """심리테스트 시뮬레이션 (랜덤 답변)"""
    print("=== 심리테스트 진행 중 ===")
    
    # 척도별 시뮬레이션 점수
    test_results = {
        'Big-Five': {
            'average': random.uniform(2.5, 4.5),
            'level': None
        },
        'CNFU': {
            'average': random.uniform(2.0, 4.0), 
            'level': None
        },
        'CVPA': {
            'average': random.uniform(3.0, 5.0),
            'level': None
        },
        'MSV': {
            'average': random.uniform(2.0, 4.5),
            'level': None
        },
        'SSS': {
            'average': random.uniform(2.5, 4.0),
            'level': None
        }
    }
    
    # 레벨 계산
    for scale, result in test_results.items():
        avg = result['average']
        if avg >= 4:
            result['level'] = 'High'
        elif avg >= 3:
            result['level'] = 'Medium'  
        else:
            result['level'] = 'Low'
        
        print(f"{scale}: {avg:.2f}점 ({result['level']})")
    
    return test_results

# 심리테스트 실행
psychology_results = simulate_psychology_test()
```
